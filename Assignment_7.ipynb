```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Q1: Randomized Sales Data Generation
np.random.seed(123456)  # Replace with roll number if available
sales_data = np.random.randint(1000, 5000, (12, 4))
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df_sales = pd.DataFrame(sales_data, columns=categories, index=months)

# Display first 5 rows and summary statistics
print(df_sales.head())
print(df_sales.describe())

# Total sales per category and per month
df_sales['Total Sales'] = df_sales.sum(axis=1)
print(df_sales)

# Calculate Growth Rate
df_sales['Growth Rate'] = df_sales['Total Sales'].pct_change() * 100

# Apply a randomized discount
roll_number = 123456  # Replace with actual roll number
if roll_number % 2 == 0:
    df_sales['Electronics'] *= 0.9
else:
    df_sales['Clothing'] *= 0.85

# Plot monthly sales trends
plt.figure(figsize=(10, 6))
df_sales.drop(columns=['Total Sales', 'Growth Rate']).plot(marker='o')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()

# Box plots for sales distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_sales.drop(columns=['Total Sales', 'Growth Rate']))
plt.title('Sales Distribution by Category')
plt.grid()
plt.show()

# Q2: Array Operations
array = np.array([[1, -2, 3], [-4, 5, -6]])
print("Absolute Values:", np.abs(array))
print("Percentiles:", np.percentile(array, [25, 50, 75], axis=None))
print("Mean:", np.mean(array), "Median:", np.median(array), "Std Dev:", np.std(array))

# Q3: Floor, Ceiling, Truncate, and Rounding
arr = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
print("Floor:", np.floor(arr))
print("Ceiling:", np.ceil(arr))
print("Truncated:", np.trunc(arr))
print("Rounded:", np.round(arr))

# Q4: Swap two elements in a list using a temporary variable
def swap_list_elements(lst, idx1, idx2):
    temp = lst[idx1]
    lst[idx1] = lst[idx2]
    lst[idx2] = temp
    return lst

my_list = [1, 2, 3, 4, 5]
print("Swapped List:", swap_list_elements(my_list, 1, 3))

# Q5: Swap two elements in a set by converting to a list
def swap_set_elements(my_set, val1, val2):
    temp_list = list(my_set)
    idx1, idx2 = temp_list.index(val1), temp_list.index(val2)
    temp_list[idx1], temp_list[idx2] = temp_list[idx2], temp_list[idx1]
    return set(temp_list)

my_set = {10, 20, 30, 40, 50}
print("Swapped Set:", swap_set_elements(my_set, 20, 40))
```
